```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("lending_club_loan_two.csv")
display(df.head())
```

```python
temp = df.isna().sum() 
display(temp[temp > 0])
```

```python
df.info()
```

```python
df['issue_d'] = pd.to_datetime(df['issue_d'])
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'])

df['issue_year'] = df['issue_d'].dt.year
df['issue_month'] = df['issue_d'].dt.month

df['cr_year'] = df['earliest_cr_line'].dt.year
df['cr_month'] = df['earliest_cr_line'].dt.month
```

# Class Imbalance

Fully Paid :=  Didn't Default  
Charged off := Defaulted

```python
target_col = 'loan_status'
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x = df[target_col])
```

## Relation of Loan_Status with Categorical Columns

```python
def plot_count(data, x, hue = target_col, figsize = (3,3), ax = None, order = None):
    if ax is None:
        plt.figure(figsize=figsize)
        sns.countplot(data = df, x = x, hue = target_col, order = order)
        plt.show()
    else:
        sns.countplot(data = df, x = x, hue = target_col, ax = ax, order = order)
```

```python
plot_count(df, 'term')
```

```python
_, ax = plt.subplots(1,2, figsize=(20,4))
plot_count(df, x = 'grade',  ax = ax[0], order = sorted(set(df['grade'])))
plot_count(df, x = 'sub_grade', ax = ax[1], order = sorted(set(df['sub_grade'])))
```

## Employement

```python
print('Number of unique employement titles : ', len(df['emp_title'].unique()))
```

```python
top_ = 50
emp_titles = df['emp_title'].value_counts().index[:top_].to_list()
print(f'Top {top_} profession:',emp_titles)
plt.figure(figsize=(12,3))
t_ = df[df['emp_title'].isin(emp_titles)]
sns.countplot(data = t_, x = 'emp_title', hue = target_col, order = emp_titles)
plt.xticks(rotation=90)
plt.show()
```

```python
emp_length_order = ['< 1 year', '1 year', '2 years', '3 years', '4 years', '5 years', '6 years', '7 years',
                    '8 years', '9 years', '10+ years']
plot_count(df, 'emp_length', figsize=(12,3), order = emp_length_order)
```

## Credit Line

```python
t_ = df[['cr_year','cr_month', 'loan_status', 'earliest_cr_line']].groupby(['cr_year','cr_month', 'loan_status']).count()
t_.columns = ['count']
t_ = t_.reset_index()
plt.figure(figsize=(12,4))
sns.lineplot(data = t_, x = 'cr_year', y='count', hue='loan_status', marker="o")
plt.xticks(sorted(set(df['cr_year'])), rotation=90)
plt.show()
```

## Issued Year

```python
t_ = df[['issue_year','issue_month', 'loan_status', 'issue_d']].groupby(['issue_year','issue_month', 'loan_status']).count()
t_.columns = ['count']
t_ = t_.reset_index()
plt.figure(figsize=(12,4))
sns.lineplot(data = t_, x = 'issue_year', y='count', hue='loan_status', marker="o")
plt.xticks(sorted(set(df['issue_year'])))
plt.show()
```

### Patterns for grouped years :
* 2008 - 2009 : Years of recession
* 2009 - 2013 : Steady Growth after recession 
* 2014 - 2015 : Strange monthly patterns appearing in both the years such as drop in 9, increase in 10 

```python
_, ax = plt.subplots(10, 1, figsize=(30,60))
for idx, year in enumerate(range(2007, 2017)):
    sns.lineplot(t_[t_['issue_year']==year], x = 'issue_month', y='count', hue = 'loan_status',
                 marker = 'o',
                 ax = ax[idx] )
    ax[idx].set_xticks([i for i in range(1,13)])
    ax[idx].set_title(year)
```
